```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
import nltk

# download resource tokenizer
nltk.download('punkt')

# baca file csv
df = pd.read_csv('youtube_comments.csv')

# tampilkan 5 baris pertama
df.head()
```

```python
def clean_text(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)  # hapus mention
    text = re.sub(r'[^A-Za-zÀ-ÿ\s]', '', text)  # hapus tanda baca & angka
    text = text.lower()  # ubah ke huruf kecil
    text = text.strip()  # hapus spasi berlebih
    return text

df['clean_comment'] = df['comment'].astype(str).apply(clean_text)
df.head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')  # <--- tambahkan baris ini

from nltk.tokenize import word_tokenize

df['tokens'] = df['clean_comment'].apply(word_tokenize)
df[['comment', 'tokens']].head()
```

```python
df.to_csv('youtube_comments_tokenized.csv', index=False)
from google.colab import files
files.download('youtube_comments_tokenized.csv')
```
