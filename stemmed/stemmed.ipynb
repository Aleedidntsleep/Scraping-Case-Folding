```python
!pip install Sastrawi

import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Baca file hasil filtering
df = pd.read_csv('youtube_comments_filtered.csv')

# Inisialisasi stemmer Bahasa Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Lakukan stemming pada kolom filtered_comment
df['stemmed_comment'] = df['filtered_comment'].astype(str).apply(stemmer.stem)

# Tampilkan hasil awal
df[['filtered_comment', 'stemmed_comment']].head()
```
